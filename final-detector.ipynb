```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import os

# ------------------------
# GPU check (no memory growth)
# ------------------------
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print("GPU ready:", gpus[0])
else:
    print("Using CPU")

# ------------------------
# Data pipeline
# ------------------------
train_dir = '/kaggle/input/datasets/birdy654/cifake-real-and-ai-generated-synthetic-images/train'
test_dir  = '/kaggle/input/datasets/birdy654/cifake-real-and-ai-generated-synthetic-images/test'

datagen = ImageDataGenerator(rescale=1./255)

train_gen = datagen.flow_from_directory(
    train_dir,
    target_size=(32,32),
    batch_size=32,
    class_mode='binary'
)

test_gen = datagen.flow_from_directory(
    test_dir,
    target_size=(32,32),
    batch_size=32,
    class_mode='binary',
    shuffle=False
)

# ------------------------
# Tiny CNN
# ------------------------
model = Sequential([
    Conv2D(16,(3,3),activation='relu',input_shape=(32,32,3)),
    MaxPooling2D(2,2),
    Conv2D(32,(3,3),activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(64,activation='relu'),
    Dropout(0.3),
    Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()

# ------------------------
# Train
# ------------------------
model.fit(train_gen, validation_data=test_gen, epochs=5)

# ------------------------
# Predict
# ------------------------
preds = model.predict(test_gen)
print(preds[:10])

# ------------------------
# Save trained model
# ------------------------
MODEL_DIR = '/kaggle/working/models'
os.makedirs(MODEL_DIR, exist_ok=True)
MODEL_PATH = os.path.join(MODEL_DIR, 'my_model.h5')

model.save(MODEL_PATH)
print(f"Model saved at: {MODEL_PATH}")
```

```python
import os
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model

# ------------------------
# Path and params
# ------------------------
TEST_DIR = "/kaggle/input/datasets/birdy654/cifake-real-and-ai-generated-synthetic-images/test/FAKE"  # your folder here
IMG_SIZE = (32, 32)  # same as your CNN input

# ------------------------
# Load the saved model
# ------------------------
model = load_model('/kaggle/working/models/my_model.h5')

# ------------------------
# Preprocess function
# ------------------------
def preprocess_img(img_path):
    img = image.load_img(img_path, target_size=IMG_SIZE)
    img_array = image.img_to_array(img) / 255.0
    img_array = np.expand_dims(img_array, axis=0)
    return img_array

# ------------------------
# Predict all images in the folder
# ------------------------
test_files = [f for f in os.listdir(TEST_DIR) if os.path.isfile(os.path.join(TEST_DIR, f))]

for img_file in test_files:
    img_path = os.path.join(TEST_DIR, img_file)
    img_input = preprocess_img(img_path)
    pred = model.predict(img_input)[0][0]  # sigmoid output
    label = "REAL" if pred > 0.5 else "AI"
    print(f"{img_file}: {label} ({pred:.5f})")
```
